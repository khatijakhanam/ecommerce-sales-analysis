```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/cleaned_data.csv')

# Monthly revenue trend
monthly = df.groupby('month')['revenue'].sum()
plt.figure(figsize=(10,4))
plt.plot(monthly.index, monthly.values, marker='o', color='steelblue')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month'); plt.ylabel('Revenue ($)')
plt.savefig('../data/monthly_trend.png')
plt.show()

# Top products
top_products = df.groupby('product')['revenue'].sum().nlargest(5)
top_products.plot(kind='barh', color='purple')
plt.title('Top 5 Products by Revenue')
plt.savefig('../data/top_products.png')
plt.show()

# Category split
df.groupby('category')['revenue'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Revenue by Category')
plt.savefig('../data/category_pie.png')
plt.show()
```
